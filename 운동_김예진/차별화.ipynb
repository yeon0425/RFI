## 서울시 운동을 하지 않는 이유 통계
1. 운동을 할 충분한 시간이 없어서
2. 운동을 싫어해서

운동 방법을 모른다기보다 시간적/ 심리적 여유가 없다고 볼 수 있음.

**이에 기반하여 프로젝트 목표 정하기**
- 많은 시간이 소요되지 않도록.
- 부담스럽지 않고 쉽게 할 수 있는 운동
- **`Input`** : 운동할 수 있는 시간, 선호하는 운동 (요가,걷기 등 간단한 운동/ 격렬한 운동), 사용 가능한 장비, 목표 (체중 감량, 체력 강화 등)

#### 간단한 예시 프로그램 ..

```python
# 사용자 프로필 입력 받기
time_available = int(input("하루에 가능한 운동 시간은 몇 분인지 (ex. 30) : "))
exercise_preferences = input("선호하는 운동 종류 (ex. 요가, 걷기, 스쿼트) : ")
available_equipment = input("운동기구 소유 여부 (y/n) : ")
goals = input("운동을 하는 이유 (ex. 다이어트, 근육 강화, 스트레스 해소) : ")

# 운동 추천 알고리즘
if time_available <= 30:
    if "요가" in exercise_preferences:
        workout = "10분 요가"
    elif "걷기" in exercise_preferences:
        workout = "30분 빠른 걷기"
    else:
        workout = "10분 스쿼트"
elif time_available <= 60:
    if "요가" in exercise_preferences:
        workout = "30분 요가"
    elif "걷기" in exercise_preferences:
        workout = "60분 빠른 걷기"
    else:
        workout = "30분 스쿼트"
else:
    if "요가" in exercise_preferences:
        workout = "60분 요가"
    elif "걷기" in exercise_preferences:
        workout = "90분 빠른 걷기"
    else:
        workout = "45분 스쿼트"

if "y" in available_equipment:
    equipment = "덤벨"
else:
    equipment = "바디웨이트"

# 사용자에게 추천 운동 제안
print("추천 운동:", workout, "(",equipment, "사용)") 
```
